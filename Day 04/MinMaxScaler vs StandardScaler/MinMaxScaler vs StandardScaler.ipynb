```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title("Original Feature Distributions (Unscaled)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
minmax = MinMaxScaler()
standard = StandardScaler()
```

```python
df_minmax = pd.DataFrame(minmax.fit_transform(df), columns=df.columns)
df_minmax.head()
```

```python
df_standard = pd.DataFrame(standard.fit_transform(df), columns=df.columns)
df_standard.head()
```

```python
# Plot side-by-side boxplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(data=df_minmax, ax=axs[0])
axs[0].set_title("MinMax Scaled Features")
axs[0].tick_params(axis='x', rotation=45)
axs[0].grid(True)

sns.boxplot(data=df_standard, ax=axs[1])
axs[1].set_title("Standard Scaled Features")
axs[1].tick_params(axis='x', rotation=45)
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
