```python
from sklearn.model_selection import learning_curve, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np

# 1. Load bigger dataset
X, y = load_digits(return_X_y=True)
model = LogisticRegression(max_iter=500)  # more iterations for convergence

# 2. Define stratified KFold
cv = StratifiedKFold(n_splits=5)

# 3. Get learning curve data
train_sizes, train_scores, valid_scores = learning_curve(
    model, X, y, cv=cv, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 5)  # 10%, 32%, 55%, 77%, 100%
)

# 4. Calculate means
train_scores_mean = np.mean(train_scores, axis=1)
valid_scores_mean = np.mean(valid_scores, axis=1)

# 5. Plot
plt.plot(train_sizes, train_scores_mean, label="Training score")
plt.plot(train_sizes, valid_scores_mean, label="Validation score")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve (Digits Dataset)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression

# 1. Load dataset
X, y = load_digits(return_X_y=True)

# 2. Create model
model = LogisticRegression(max_iter=500)

# 3. Choose a hyperparameter to vary: C (inverse of regularization strength)
param_range = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

# 4. Get validation curve data
train_scores, valid_scores = validation_curve(
    model, X, y, 
    param_name="C", 
    param_range=param_range,
    cv=5,
    scoring="accuracy"
)

# 5. Calculate means
train_scores_mean = np.mean(train_scores, axis=1)
valid_scores_mean = np.mean(valid_scores, axis=1)

# 6. Plot
plt.plot(param_range, train_scores_mean, label="Training score")
plt.plot(param_range, valid_scores_mean, label="Validation score")
plt.xscale('log')  # because C values are log-scale
plt.xlabel("C (Inverse Regularization Strength)")
plt.ylabel("Accuracy")
plt.title("Validation Curve for Logistic Regression")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier

# 1. Load dataset
X, y = load_digits(return_X_y=True)

# 2. Create model
model = KNeighborsClassifier(max_iter=500)

# 3. Choose a hyperparameter to vary: C (inverse of regularization strength)
param_range = [1, 3, 5, 7, 9, 11, 13]

# 4. Get validation curve data
train_scores, valid_scores = validation_curve(
    model, X, y, 
    param_name="n_neighbors", 
    param_range=param_range,
    cv=5,
    scoring="accuracy"
)

# 5. Calculate means
train_scores_mean = np.mean(train_scores, axis=1)
valid_scores_mean = np.mean(valid_scores, axis=1)

# 6. Plot
plt.plot(param_range, train_scores_mean, label="Training score")
plt.plot(param_range, valid_scores_mean, label="Validation score")
plt.xscale('log')  # because C values are log-scale
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.title("Validation Curve for neighbors")
plt.legend()
plt.grid(True)
plt.show()
```
